```python
# Define the function
def my_function(x):
    return x**2 + x + 1

# Define your own numerical derivative function
def derivative(func, x0, dx=1e-6):
    return (func(x0 + dx) - func(x0 - dx)) / (2 * dx)

# Calculate the derivative at x = 2
result = derivative(func=my_function, x0=2)
print("Derivative at x=2:", result)
```

```python
# define the function
def my_function(x):
    return x**2 + x + 1

# calculate the first derivative of the function at x = 2

derivative(func= my_function, x0=2)
```

#### Function: `f(x) = x**2 + x + 1`
#### Derivate:  `f'(x) = 2*x + 1`
#### Solution:  `f'(2) = 2*2 + 1 = 5`

---

***Now, calculate the second derivative***

---

```python
# Define the function
def my_function(x):
    return x**2 + x + 1

# Define a numerical derivative function supporting n=1 or n=2
def derivative(func, x0, dx=1e-6, n=1):
    if n == 1:
        return (func(x0 + dx) - func(x0 - dx)) / (2 * dx)
    elif n == 2:
        return (func(x0 + dx) - 2*func(x0) + func(x0 - dx)) / (dx**2)
    else:
        raise ValueError("Only n=1 and n=2 are supported")

# First derivative at x=2
print("1st derivative at x=2:", derivative(func=my_function, x0=2, n=1))

# Second derivative at x=2
print("2nd derivative at x=2:", derivative(func=my_function, x0=2, n=2))
```

```python
from sklearn.datasets import fetch_california_housing
import pandas as pd

# Load dataset
housing = fetch_california_housing(as_frame=True)
data = housing.frame  # entire dataset as DataFrame

# Separate features and target
features = data.drop(columns=['MedHouseVal'])
target = data['MedHouseVal']

# Show first 5 rows of features
features.head()
```

---

#### `SEPARATE THE TARGET VARIABLE`

---

```python
# Features (all except target)
features = data.drop(columns=['MedHouseVal'])

# Target (as a separate DataFrame)
target = data[['MedHouseVal']]  # keep as DataFrame
```

---

#### `CREATE A LINEAR REGRESSION MODEL`

---

```python
# Features
X = features  # your features DataFrame

# Target
y = target['MedHouseVal']  # the correct column name
```

---

#### `BUILD A MODEL`


---

```python
import statsmodels.api as sm

# Add intercept (constant term) to features
X_const = sm.add_constant(X)

# Fit the OLS model
model = sm.OLS(y, X_const).fit()

# View the summary
print(model.summary())
```

---

#### `MAKE PREDICTIONS`

---

```python
# Make sure to add the constant to X before predicting
X_const = sm.add_constant(X)  # same as used for fitting

predictions = model.predict(X_const)

# Check the first few predictions
print(predictions.head())
```

```python
# open and load the data in json file
with open('C:\Arun\Data/nested.json') as f :
    my_json_data = json.load(f)

print(my_json_data)
```

```python
# use pprint or (pretty print) to print the data in the structured format
from pprint import pprint
pprint(my_json_data)
```

```python
# We saw that data in the file is in json list form.
# iterate through json 

data_0 = my_json_data[0]
data_0
```

```python
data_0['details']
```

```python
data_0['details']['age']
```

```python
# iterate through the json data
for data in my_json_data:
    print(data['details']['age'])
```

```python
# create a new empty list to store the filtered data
filtered_data = []

# iterate through the json data
for data in my_json_data:
    
    # create new empty dictionary
    filtered_variable = {}
    
    # check for the condition
    if data['details']['age'] > 15:
        # if condition satisfies, store the age and name
        filtered_variable['age'] = data['details']['age']
        filtered_variable['name']= data['details']['name']
        filtered_data.append(filtered_variable)
```

```python
# check the filtered data
filtered_data
```

```python
# put the filtered data into the new json file
with open('C:\Arun\Data/filtered.json','w') as f:
    
    json.dump(filtered_data, f, indent=4)
```

# Exercise: RGB intensity plot

```python
import numpy as np
import matplotlib.pyplot as plt

# The user's code to be filled in.
def plot_intensity(image, row):
    """
    Plots the intensity of each RGB channel for a given row of an image.

    Args:
        image (numpy.ndarray): The input image (shape: H, W, 3).
        row (int): The row index to plot.
    """
    # Fill in the three lines below.
    # We select the specified 'row' from the image.
    # The image is a 3D array (height, width, channels).
    # image[row] gives us all pixels in that row, which is a (width, 3) array.
    # We then use slicing to get all columns of that row for each specific channel.
    red_values = image[row, :, 0]
    green_values = image[row, :, 1]
    blue_values = image[row, :, 2]

    plt.figure()
    plt.plot(red_values, color='red', label='Red Channel')
    plt.plot(green_values, color='green', label='Green Channel')
    plt.plot(blue_values, color='blue', label='Blue Channel')

    # Adding labels and a title for better visualization.
    plt.title(f'RGB Intensity Plot for Row {row}')
    plt.xlabel('Column')
    plt.ylabel('Intensity')
    plt.legend()
    plt.grid(True)
    
    plt.show()

# Test the function here.
# Note: Since the image is not provided, we will create a dummy image for demonstration.
# In a real scenario, you would load an image, e.g., using skimage or imageio.
from skimage.data import chelsea

cat = chelsea()

plot_intensity(cat, 50)
plot_intensity(cat, 100)
```

# Exercise: Convert to Black and White

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import data, img_as_float, color

# Load the 'chelsea' image from scikit-image data
cat = data.chelsea()

# Convert the RGB image to a grayscale image using the standard
# luminance-preserving conversion formula.
# The '@' operator in this context is for matrix multiplication in NumPy.
# Here, it performs a dot product of the RGB channels with the
# luminance coefficients.
cat_gray = cat @ [0.2126, 0.7152, 0.0722]

# Create a figure with two subplots side by side
f, (ax0, ax1) = plt.subplots(1, 2)

# Display the manually calculated grayscale image on the first subplot
ax0.imshow(cat_gray, cmap='gray')

# Display the grayscale image converted using scikit-image's built-in function
# on the second subplot.
ax1.imshow(color.rgb2gray(cat), cmap='gray');

# Let's calculate the mean intensity error, as compared to
# scikit-image's color.rgb2gray
# The manual conversion returns an integer array, while `rgb2gray` returns a float
# array scaled from 0.0 to 1.0. We must normalize our manual conversion by 255
# to make a fair comparison.
mean_error = np.mean(np.abs((cat_gray / 255) - color.rgb2gray(cat)))

# Print the mean error
print(f"Mean intensity error: {mean_error}")

# Display the plots
plt.show()
```

# RANSAC (random sample consenuse)

```python
import numpy as np
from matplotlib import pyplot as plt
from skimage.measure import ransac
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(seed=1)

# generate coordinates of line
x = np.arange(-200, 200)
y = 0.2 * x + 20
data = np.column_stack([x, y])

# add faulty data
faulty = np.array(30 * [(180., -100)])
faulty += 5 * np.random.normal(size=faulty.shape)
data[:faulty.shape[0]] = faulty

# add gaussian noise to coordinates
noise = np.random.normal(size=data.shape)
data += 0.5 * noise
data[::2] += 5 * noise[::2]
data[::4] += 20 * noise[::4]

plt.plot(data[:, 0], data[:, 1], 'b.')
plt.show() # added plt.show() for proper display
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Assuming 'data' is already defined from a previous step
# For example: data = np.column_stack([x_values, y_values])

# Separate the x and y coordinates from the data
x = data[:, 0]
y = data[:, 1]

# Construct the design matrix X for linear regression
# The first column is the x values, and the second is a column of ones
# This setup is for the equation y = mx + c, where the ones
# correspond to the intercept 'c'
X = np.column_stack((x, np.ones_like(x)))

# Use NumPy's least squares solver to find the parameters 'p'
# p will contain the slope (m) and the y-intercept (c)
p, _, _, _ = np.linalg.lstsq(X, y, rcond=None)
p
```

```python
# Unpack the parameters into 'm' (slope) and 'c' (intercept)
m, c = p

# Plot the original data points as blue dots
plt.plot(x, y, 'b.')

# Create a new set of x values for the fitted line
xx = np.arange(-250, 250)

# Calculate the corresponding y values for the fitted line
yy = m * xx + c

# Plot the fitted line in red
plt.plot(xx, yy, 'r-');

plt.show()
```

```python
from skimage.measure import ransac, LineModelND
import numpy as np
import matplotlib.pyplot as plt

# --- Part 1: Initial Linear Model Estimation ---
# Assumes 'data', 'x', 'y', and 'xx' are defined from a previous step
# E.g., from the linear regression code snippet provided previously

# Create an instance of the linear model
model = LineModelND()

# Estimate the model parameters from all the data points
model.estimate(data)

# Unpack the model parameters to get the origin and direction
origin, direction = model.params
```

```python
# Plot the original data points
plt.plot(x, y, 'b.')

# Plot the line predicted by the simple linear model
plt.plot(xx, model.predict_y(xx), 'r-');

plt.title('Initial Linear Model')
plt.show()
```

```python
# Use the RANSAC algorithm to find a robust linear model
# It will automatically detect and ignore outliers
model_robust, inliers = ransac(data, LineModelND, min_samples=2,
                               residual_threshold=10, max_trials=1000)

# Create a boolean array to identify the outliers
outliers = (inliers == False)

# Predict the y-values for the robust model
yy = model_robust.predict_y(xx)

# Create a new figure for plotting the RANSAC results
fig, ax = plt.subplots()

# Plot the 'inlier' data points in blue
ax.plot(data[inliers, 0], data[inliers, 1], 'b.', alpha=0.6, label='Inlier data')

# Plot the 'outlier' data points in red
ax.plot(data[outliers, 0], data[outliers, 1], '.r', alpha=0.6, label='Outlier data')

# Plot the robust linear model line
ax.plot(xx, yy, '-b', label='Robust line model')

# Add a legend to the plot
plt.legend(loc='lower left')

plt.title('RANSAC Robust Linear Model')
plt.show()
```
